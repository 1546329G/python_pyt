combinar y agrupar datos en pandas

```python
import pandas as pd
df1= pd.DataFrame({
    'producto': ['manzana', 'Platano', 'Mandarina'],
    'Enero': [30,40,20],
    'Febrero':[35,50,25]

})

df2= pd.DataFrame({
    'producto': ['manzana', 'Platano', 'Mandarina'],
    'Marzo': [25,55,30],
    'Abril':[40,60,35]
})
#combinar los data frame
ventas=pd.merge(df1,df2, on='producto')
#calcular las ventas totales
ventas['total']=ventas[['Enero','Febrero','Marzo','Abril']].sum(axis=1)
print(ventas)

#2
ventas1=pd.merge(df1,df2, on='producto')
ventasFila=ventas1[['Enero','Febrero','Marzo','Abril']].sum(axis=0)

print(ventasFila)
```

```python
#aplicar funciunes  columnas
import pandas as pd
import numpy as np

#crear  dataframe
data=np.random.randint(1,100,size=(5,4))
df=pd.DataFrame(data, columns=['a','b','c','d'])

#Aplicar una funcion lamba  para normalizar los datos
df_normalizado = df.apply(lambda x:(x-x.min())/(x.max()-x.min()))
print('DataFrame Normalizacion\n',df_normalizado)
#la normalizacion es un procso de que ajista los valorees para que sten en un
#rango comun,
```

```python
# creamosdata frame con vbalores faltantes
import pandas as pd
data={
    'profucto' :['arandano','pepinillo','durazno'],
    'precio' : [5.6,np.nan,5.3],
    'Stock' :  [200,250,np.nan]
}

df=pd.DataFrame(data)
#rellena los valoes faltantes con los producros de cada columna
df['precio'].fillna(df['precio'].mean(),inplace=True)
df['Stock'].fillna(df['Stock'].mean(),inplace=True)
df
```
